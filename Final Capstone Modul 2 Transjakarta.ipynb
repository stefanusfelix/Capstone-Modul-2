## **Dataset Understanding**
1.	transID: Unique transaction id for every transaction
2.	payCardID: Customers main identifier. The card customers use as a ticket for entrance and exit.
3.	payCardBank: Customers card bank issuer name
4.	payCardName: Customers name that is embedded in the card.
5.	payCardSex: Customers sex that is embedded in the card
6.	payCardBirthDate: Customers birth year
7.	corridorID: Corridor ID / Route ID as key for route grouping.
8.	corridorName: Corridor Name / Route Name contains Start and Finish for each route.
9.	direction: 0 for Go, 1 for Back. Direction of the route.
10.	tapInStops: Tap In (entrance) Stops ID for identifying stops name
11.	tapInStopsName: Tap In (entrance) Stops Name where customers tap in.
12.	tapInStopsLat: Latitude of Tap In Stops
13.	tapInStopsLon: Longitude of Tap In Stops
14.	stopStartSeq: Sequence of the stops, 1st stop, 2nd stops etc. Related to direction.
15.	tapInTime: Time of tap in. Date and time
16.	tapOutStops: Tap Out (Exit) Stops ID for identifying stops name
17.	tapOutStopsName: Tap out (exit) Stops Name where customers tap out.
18.	tapOutStopsLat: Latitude of Tap Out Stops
19.	tapOutStopsLon: Longitude of Tap Out Stops
20.	stopEndSeq: Sequence of the stops, 1st stop, 2nd stops etc. Related to direction.
21.	tapOutTime: Time of tap out. Date and time
22.	payAmount: The number of what customers pay. Some are free. Some not.


## **Company Profile: PT Transportasi Jakarta**
PT Transportasi Jakarta (Transjakarta) adalah Badan Usaha Milik Daerah (BUMD) yang mengoperasikan layanan angkutan umum berbasis bus rapid transit (BRT) di Jakarta. Didirikan pada tahun 2004, Transjakarta bertujuan untuk meningkatkan mobilitas penduduk Jakarta, mengurangi kemacetan, dan mengurangi emisi gas buang. Layanan Transjakarta terdiri dari rute-rute bus yang melintasi kota dengan jalur khusus, berhenti di stasiun-stasiun yang telah ditentukan, dan menggunakan sistem pembayaran non-tunai dengan kartu elektronik. 

PT Transjakarta menyediakan 3 jenis layanan:
1. Layanan premium, Royaltrans, harga sebesar Rp 20.000
2. Layanan reguler Transjakarta, harga sebesar Rp 3.500
3. Layanan Mikrotrans, harga sebesar 0 atau gratis

## **Main Problems**

![overcapacity](https://statik.tempo.co/data/2013/01/01/id_159069/159069_620.jpg)

Berdasarkan artikel dari KOMPAS.com, Pengurus Harian Yayasan Lembaga Konsumen Indonesia (YLKI) Sri Wahyuni mengatakan, kelebihan penumpang dalam bus (over capacity) menjadi salah satu hal yang paling banyak dikeluhkan penumpang transjakarta. (https://megapolitan.kompas.com/read/2017/09/14/14491311/ylki-keluhan-terbanyak-soal-transjakarta-adalah-over-capacity?page=all.)
- Kendala overcapacity: Overcapacity dalam transportasi publik menjadi kendala utama untuk mendorong masyarakat beralih dari kendaraan pribadi ke transportasi umum.

Berdasarkan artikel dari Republik Merdeka (https://rmol.id/nusantara/read/2023/09/19/589686/over-kapasitas-bikin-masyarakat-malas-naik-transportasi-publik-di-jakarta), permasalahan overcapacity memberikan beberapa dampak negatif, sebagai berikut:
- Menurunkan minat masyarakat: Jumlah pengguna layanan transportasi publik di Jakarta saat ini mencapai 1,8 juta penumpang per hari, namun hanya sekitar 9,6% dari jumlah penduduk.
- Meningkatkan dominasi kendaraan pribadi: Penggunaan kendaraan pribadi, terutama motor pribadi, masih mendominasi transportasi di Jakarta dengan persentase sebesar 60%, diikuti oleh mobil pribadi sebanyak 20%.

Berdasarkan artikel dari Republik Merdeka (https://rmol.id/nusantara/read/2023/09/19/589686/over-kapasitas-bikin-masyarakat-malas-naik-transportasi-publik-di-jakarta), selain disebabkan oleh overcapacity, terdapat beberapa penyebab lainnya, seperti:
- Kendala pengguna: Beberapa isu utama yang dihadapi oleh pengguna transportasi publik yaitu kenyamanan, keandalan, dan kebutuhan untuk berpindah rute atau moda transportasi untuk mencapai tujuan.
- Kendala non-pengguna: Faktor-faktor seperti kenyamanan, waktu tempuh yang lama, dan keandalan juga menjadi hambatan bagi non-pengguna transportasi publik untuk beralih ke transportasi publik.

## **Problems**

**Overcapacity vs Undercapacity**
- Overcapacity saat peak hours (jam berangkat dan pulang kerja)
- Undercapacity saat di luar jam peak hours

Dampak dari overcapacity:
- Risiko pada keselamatan penumpang.
- Menurunkan minat masyarakat untuk beralih ke transportasi umum.
- Menurunkan masa pakai kendaraan umum.
- Keterlambatan jadwal.

Dampak dari undercapacity:
- Meningkatnya operational cost 

## **Solution Recommendation**

**1. Transport Resource Management**
- Fokus pada pengalokasian jumlah armada secara tepat. Dalam hal ini, PT Transportasi Jakarta dapat fokus untuk menyesuaikan jumlah armada pada corridor yang ramai saat peak hours. Hal ini perlu didukung dengan Intelligent Transportation System.
- Melakukan perawatan armada secara berkala. Pada saat jam di luar peak hours, dapat mengurangi jumlah armada yang digunakan. Melakukan perawatan pada beberapa armada yang tidak digunakan dan merotasi armada yang sudah diservis dan belum diservis.
- Fokus pada pemberdayaan sumber daya manusia. Tanpa ada sumber daya manusia yang berkualitas, kegiatan operasional dan layanan kepada customer tidak akan maksimal.

**2. Intelligent Transport System**
- Membuat one integrated system application yang terkoneksi dengan layanan Transjakarta, Microtrans, dan Royaltrans. 
- Pengalokasian jumlah armada yang sesuai dapat dilakukan dengan memberlakukan sistem pemesanan. Penumpang yang melakukan pemesanan melalui aplikasi akan diprioritaskan terlebih dahulu.
- Sistem pemesanan: 
1. Calon penumpang akan memesan jadwal rute dan lokasi tap in dan tap out (lokasi penjemputan dan lokasi tujuan). Pemesanan dapat dilakukan sehari sebelumnya.
2. Calon penumpang akan mendapatkan notifikasi untuk melakukan konfirmasi (2 jam sebelum jadwal keberangkatan) dan akan mendapatkan barcode. Apabila 1 jam sebelum jadwal keberangkatan, calon penumpang tidak melakukan konfirmasi, maka tidak akan memperoleh barcode. Barcode akan digunakan saat calon penumpang akan menaiki armada. Dengan ini, PT Transjakarta dapat mengatur jumlah armada dan jadwal yang sesuai dengan tingkat permintaan customer. 
3. Sistem pembayaran akan terintegrasi ke dalam aplikasi. Sehingga, yang sebelumnya menggunakan kartu fisik, seperti kartu Jakcard (dki), e-money(mandiri), flazz (bca) dapat berubah menjadi kartu pembayaran digital. Dengan sistem integrasi ini, yang sebelumnya, penumpang melakukan scan kartu pembayaran fisik saat menaiki armada berubah menjadi scan barcode yang telah terkoneksi dengan kartu pembayaran digital.
4. Aplikasi berisikan informasi terkait rute, jadwal keberangkatan, estimasi waktu tunggu jadwal berikutnya, lokasi corridor/halte terdekat, rekomendasi corridor/halte berdasarkan lokasi penjemputan dan lokasi tujuan.
- Melakukan sosialisasi dan edukasi terkait penerapan one integrated system application dan sistem pemesanan kepada masyarakat.


**3. Transport Traffic Management**
- Fokus untuk memaksimalkan jalur khusus Transjakarta (layanan BRT) dengan menindaktegas kendaraan pribadi yang masuk jalur khusus Transjakarta.
- Fokus mengurangi kemacetan pada jalur umum (layanan non-BRT).
- Bertujuan agar jadwal yang telah direncanakan tidak mengalami delay.

**4. Transport Services Management**
- Menyediakan corridor atau halte yang layak dan mampu memenuhi kapsitas saat peak hours.
- Menyediakan armada yang layak dan nyaman bagi customer.
- Menyediakan cctv dan petugas guna menjaga tata tertib dan keamaan. Hal ini bertujuan untuk memitigasi terjadinya risiko pencopetan atau pelecehan seksual.
